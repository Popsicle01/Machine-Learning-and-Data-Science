##Cleaning and Preprocessing Data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler

# 1. Handle missing values (e.g., interpolate missing sample values in dynamic and static datasets)
DF_DA_AVG_DYNAMIC_PS_AROUSAL.fillna(method='ffill', inplace=True)
DF_DA_AVG_DYNAMIC_PS_VALENCE.fillna(method='ffill', inplace=True)

# 2. Standardize column names (convert to lowercase, replace spaces with underscores)
def standardize_column_names(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    return df

DF_DA_AVG_DYNAMIC_PS_AROUSAL = standardize_column_names(DF_DA_AVG_DYNAMIC_PS_AROUSAL)
DF_DA_AVG_DYNAMIC_PS_VALENCE = standardize_column_names(DF_DA_AVG_DYNAMIC_PS_VALENCE)
DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000 = standardize_column_names(DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000)
DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058 = standardize_column_names(DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058)

# 3. Check for duplicates and remove them
DF_DA_AVG_DYNAMIC_PS_AROUSAL.drop_duplicates(inplace=True)
DF_DA_AVG_DYNAMIC_PS_VALENCE.drop_duplicates(inplace=True)
DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000.drop_duplicates(inplace=True)
DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058.drop_duplicates(inplace=True)

# 4. Use RobustScaler - This will preserve the sign and handle outliers better
scaler = RobustScaler()

# Function to scale data while preserving negatives
def scale_preserve_sign(df, columns):
    # Store original signs
    signs = np.sign(df[columns])
    # Scale absolute values
    scaled_data = scaler.fit_transform(np.abs(df[columns]))
    # Restore signs
    scaled_data = scaled_data * signs
    return pd.DataFrame(scaled_data, columns=columns, index=df.index)

# Scale dynamic sample columns for arousal and valence
dynamic_sample_cols_arousal = [col for col in DF_DA_AVG_DYNAMIC_PS_AROUSAL.columns if "sample" in col]
DF_DA_AVG_DYNAMIC_PS_AROUSAL[dynamic_sample_cols_arousal] = scale_preserve_sign(
    DF_DA_AVG_DYNAMIC_PS_AROUSAL, dynamic_sample_cols_arousal
)

dynamic_sample_cols_valence = [col for col in DF_DA_AVG_DYNAMIC_PS_VALENCE.columns if "sample" in col]
DF_DA_AVG_DYNAMIC_PS_VALENCE[dynamic_sample_cols_valence] = scale_preserve_sign(
    DF_DA_AVG_DYNAMIC_PS_VALENCE, dynamic_sample_cols_valence
)

# Scale static values
static_valence_cols = ['_valence_mean', '_valence_std', '_arousal_mean', '_arousal_std']
DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000[static_valence_cols] = scale_preserve_sign(
    DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000, static_valence_cols
)

static_cols = ['_valence_mean', '_valence_std', '_valence__max_mean',
       '_valence_max_std', '_valence_min_mean', '_valence_min_std',
       '_arousal_mean', '_arousal_std', '_arousal_max_mean',
       '_arousal_max_std', '_arousal_min_mean', '_arousal_min_std']
DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058[static_cols] = scale_preserve_sign(
    DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058, static_cols
)

# 5. Handle outliers while preserving negative values
def cap_outliers_preserve_sign(df, cols, lower_quantile=0.01, upper_quantile=0.99):
    for col in cols:
        values = df[col]
        signs = np.sign(values)
        abs_values = np.abs(values)

        lower_cap = abs_values.quantile(lower_quantile)
        upper_cap = abs_values.quantile(upper_quantile)

        capped_abs_values = np.clip(abs_values, lower_cap, upper_cap)
        df[col] = capped_abs_values * signs
    return df

cap_outliers_preserve_sign(DF_DA_AVG_DYNAMIC_PS_AROUSAL, dynamic_sample_cols_arousal)
cap_outliers_preserve_sign(DF_DA_AVG_DYNAMIC_PS_VALENCE, dynamic_sample_cols_valence)
cap_outliers_preserve_sign(DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000, static_valence_cols)
cap_outliers_preserve_sign(DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058, static_cols)

# Save the processed data
output_static_path = base_path + "/managed_static2000_data.csv"
DF_DA_AVG_STATIC_PS_SONG_LEVEL_1_2000.to_csv(output_static_path, index=False)

output_static_path = base_path + "/managed_static2058_data.csv"
DF_DA_AVG_STATIC_PS_SONG_LEVEL_2000_2058.to_csv(output_static_path, index=False)

output_dynamic_arousal_path = base_path + "/managed_dynamic_arousal_data.csv"
DF_DA_AVG_DYNAMIC_PS_AROUSAL.to_csv(output_dynamic_arousal_path, index=False)

output_dynamic_valence_path = base_path + "/managed_dynamic_valence_data.csv"
DF_DA_AVG_DYNAMIC_PS_VALENCE.to_csv(output_dynamic_valence_path, index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
#merge dynamic data on song id
dynamic_data = pd.merge(dynamic_arousal, dynamic_valence, on="song_id", suffixes=("_arousal", "_valence"))

# Extract features (arousal and valence means) and song_id
features = dynamic_data[['arousal_mean', 'valence_mean']]
song_ids = dynamic_data['song_id']

# Define emotion labels based on arousal and valence combinations
def classify_emotion(arousal, valence):
    if arousal >= 0 and valence >= 0:  # High Arousal, High Valence
        return 'Happy'
    elif arousal >= 0 and valence < 0:  # High Arousal, Low Valence
        return 'Angry'
    elif arousal < 0 and valence < 0:   # Low Arousal, Low Valence
        return 'Sad'
    else:                               # Low Arousal, High Valence
        return 'Relaxed'

# Apply classification function to create labels
dynamic_data['emotion'] = features.apply(lambda row: classify_emotion(row['arousal_mean'], row['valence_mean']), axis=1)

# Separate features and labels
X = features.values
y = dynamic_data['emotion'].values

# Normalize features to the standard range [-1, 1]
scaler = MinMaxScaler(feature_range=(-1, 1))
X_scaled = scaler.fit_transform(X)

# Encode labels to numeric values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Define the neural network model
model = Sequential()

# Input layer
model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))

# Hidden layers
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))

# Output layer: 4 classes (Happy, Angry, Sad, Tender)
model.add(Dense(4, activation='softmax'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_split=0.2, verbose=1)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy: {accuracy * 100:.2f}%")

# Predict emotions on test set
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)

# Calculate prediction accuracy
prediction_accuracy = accuracy_score(y_test, y_pred_labels)
print(f"Prediction Accuracy: {prediction_accuracy * 100:.2f}%")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_labels, target_names=label_encoder.classes_))
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, RegressorMixin
import numpy as np
import pandas as pd
import joblib

class EnsembleRegressor(BaseEstimator, RegressorMixin):
    def __init__(self):
        self.rf = RandomForestRegressor(
            n_estimators=100,
            max_depth=6,
            random_state=42,
            n_jobs=-1
        )
        self.gb = GradientBoostingRegressor(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=6,
            random_state=42
        )
        self.svr = SVR(kernel='rbf', C=1.0, epsilon=0.1)
        self.lr = LinearRegression()

    def fit(self, X, y):
        self.rf.fit(X, y)
        self.gb.fit(X, y)
        self.svr.fit(X, y)

        # Use predictions from RF, GB, and SVR as features for LR
        meta_features = np.column_stack([
            self.rf.predict(X),
            self.gb.predict(X),
            self.svr.predict(X)
        ])
        self.lr.fit(meta_features, y)
        return self

    def predict(self, X):
        meta_features = np.column_stack([
            self.rf.predict(X),
            self.gb.predict(X),
            self.svr.predict(X)
        ])
        return self.lr.predict(meta_features)

# Merge dynamic data on song id
dynamic_data = pd.merge(dynamic_arousal, dynamic_valence, on="song_id", suffixes=("_arousal", "_valence"))

# Separate features and targets
features = dynamic_data.drop(columns=['song_id', 'arousal_mean', 'valence_mean'])
y_arousal = dynamic_data['arousal_mean']
y_valence = dynamic_data['valence_mean']

# Split the data for both targets
X_train, X_test, y_train_arousal, y_test_arousal, y_train_valence, y_test_valence = train_test_split(
    features, y_arousal, y_valence, test_size=0.35, random_state=42
)

# Train arousal model
print("Training arousal ensemble model...")
ensemble_pipeline_arousal = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler(feature_range=(-1, 1))),
    ('ensemble', EnsembleRegressor())
])
ensemble_pipeline_arousal.fit(X_train, y_train_arousal)
y_pred_arousal = ensemble_pipeline_arousal.predict(X_test)

# Train valence model
print("\nTraining valence ensemble model...")
ensemble_pipeline_valence = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler(feature_range=(-1, 1))),
    ('ensemble', EnsembleRegressor())
])
ensemble_pipeline_valence.fit(X_train, y_train_valence)
y_pred_valence = ensemble_pipeline_valence.predict(X_test)

def evaluate_predictions(y_true, y_pred, metric_name):
    """
    Evaluate predictions using both regression and classification metrics
    """
    # Regression metrics
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    # Convert to binary for classification metrics (above/below mean)
    threshold = np.mean(y_true)
    y_true_binary = (y_true > threshold).astype(int)
    y_pred_binary = (y_pred > threshold).astype(int)

    # Classification metrics
    accuracy = accuracy_score(y_true_binary, y_pred_binary)
    precision, recall, f1, _ = precision_recall_fscore_support(y_true_binary,
                                                             y_pred_binary,
                                                             average='binary')

    print(f"\n{metric_name} Evaluation:")
    print(f"Regression Metrics:")
    print(f"- Mean Squared Error: {mse:.4f}")
    print(f"- R Score: {r2:.4f}")
    print(f"\nClassification Metrics:")
    print(f"- Accuracy: {accuracy:.4f}")
    print(f"- Precision: {precision:.4f}")
    print(f"- Recall: {recall:.4f}")
    print(f"- F1 Score: {f1:.4f}")

    return {
        'mse': mse,
        'r2': r2,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

# Evaluate predictions
arousal_metrics = evaluate_predictions(y_test_arousal, y_pred_arousal, "Arousal")
valence_metrics = evaluate_predictions(y_test_valence, y_pred_valence, "Valence")

# Generate emotion predictions
def get_emotion(arousal, valence):
    if arousal > 0 and valence > 0:
        return "Happy"
    elif arousal > 0 and valence <= 0:
        return "Angry"
    elif arousal <= 0 and valence <= 0:
        return "Sad"
    else:  # arousal <= 0 and valence > 0
        return "Relaxed"

emotions = [get_emotion(a, v) for a, v in zip(y_pred_arousal, y_pred_valence)]
actual_emotions = [get_emotion(a, v) for a, v in zip(y_test_arousal, y_test_valence)]

# Calculate overall emotion prediction accuracy
emotion_accuracy = accuracy_score(actual_emotions, emotions)
print(f"\nEmotion Prediction Accuracy: {emotion_accuracy * 100:.2f}%")

# Print classification report for emotions
print("\nClassification Report:")
print(classification_report(actual_emotions, emotions))

# Save the models
joblib.dump(ensemble_pipeline_arousal, 'ensemble_arousal_model.joblib')
joblib.dump(ensemble_pipeline_valence, 'ensemble_valence_model.joblib')

# Create results DataFrame
results_df = pd.DataFrame({
    'Actual_Emotion': actual_emotions,
    'Predicted_Emotion': emotions,
    'Arousal_Actual': y_test_arousal,
    'Arousal_Predicted': y_pred_arousal,
    'Valence_Actual': y_test_valence,
    'Valence_Predicted': y_pred_valence
})

print("\nFirst few predictions:")
print(results_df[['Actual_Emotion', 'Predicted_Emotion']].head())
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, ClassifierMixin
import xgboost as xgb
import joblib

class EmotionEnsembleClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self):
        self.label_encoder = LabelEncoder()
        self.scaler = MinMaxScaler(feature_range=(-1, 1))

        # Define base models
        self.xgb_model = xgb.XGBClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=10,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            n_jobs=-1
        )

        self.rf_model = RandomForestClassifier(
            n_estimators=100,
            max_depth=10,
            random_state=42,
            n_jobs=-1
        )

        self.gb_model = GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.1,
            max_depth=5,
            random_state=42
        )

        self.svm_model = SVC(
            probability=True,
            kernel='rbf',
            random_state=42
        )

        # Create voting classifier
        self.ensemble = VotingClassifier(
            estimators=[
                ('xgb', self.xgb_model),
                ('rf', self.rf_model),
                ('gb', self.gb_model),
                ('svm', self.svm_model)
            ],
            voting='soft'
        )

    def fit(self, X, y):
        # Encode labels
        y_encoded = self.label_encoder.fit_transform(y)

        # Scale features
        X_scaled = self.scaler.fit_transform(X)

        # Fit ensemble
        self.ensemble.fit(X_scaled, y_encoded)
        return self

    def predict(self, X):
        X_scaled = self.scaler.transform(X)
        y_pred_encoded = self.ensemble.predict(X_scaled)
        return self.label_encoder.inverse_transform(y_pred_encoded)

    def predict_proba(self, X):
        X_scaled = self.scaler.transform(X)
        return self.ensemble.predict_proba(X_scaled)

# Function to classify emotions
def classify_emotion(arousal, valence):
    if arousal >= 0 and valence >= 0:
        return 'Happy'
    elif arousal >= 0 and valence < 0:
        return 'Angry'
    elif arousal < 0 and valence < 0:
        return 'Sad'
    else:
        return 'Relaxed'

# Main execution
def train_emotion_ensemble(static_data):
    print("Processing static dataset...")

    # Create emotion labels
    static_data['emotion'] = static_data.apply(
        lambda row: classify_emotion(row['_arousal_mean'], row['_valence_mean']),
        axis=1
    )

    # Extract features and labels
    X_static = static_data[[
        '_valence_mean', '_valence_std',
        '_arousal_mean', '_arousal_std'
    ]]
    y_static = static_data['emotion']

    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(
        X_static, y_static,
        test_size=0.2,
        random_state=42
    )

    # Create and train ensemble
    ensemble_classifier = EmotionEnsembleClassifier()
    ensemble_classifier.fit(X_train, y_train)

    # Make predictions
    y_pred = ensemble_classifier.predict(X_test)

    # Calculate and print metrics
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\nTest Accuracy: {accuracy * 100:.2f}%")
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Create results DataFrame
    results_df = pd.DataFrame({
        'Actual_Emotion': y_test,
        'Predicted_Emotion': y_pred
    })

    print("\nFirst few predictions:")
    print(results_df.head())

    # Save the model
    joblib.dump(ensemble_classifier, 'emotion_ensemble_model.joblib')

    return ensemble_classifier, results_df

# Example usage:
ensemble_classifier, results = train_emotion_ensemble(static_data)
```
